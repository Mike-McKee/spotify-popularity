# Word Cloud Fun

Here I'm just messing around with Python's `wordcloud` library to bring back these nostalgic visuals. Since I only have 439 songs, there's not a wide range of words to choose from in the track names.

But I'm not creating this wordcloud for data analysis.

I'm just doing it to have fun and learn new ways to use `wordcloud` and `matplotlib`.

```python
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
playlist = pd.read_csv('files/playlist.csv')
```

```python
tracks = []
for i in playlist['Track_Name']:
    tracks.append(i)
```

## Default Colors

First I'll creaet a simple wordcloud using the library's default colors.

```python
tracks_text = ' '.join(tracks)
wordcloud = WordCloud().generate(tracks_text)

wordcloud = WordCloud(max_font_size=50).generate(tracks_text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

## Adding Some Flair

Now I'm creating a wordcloud using my own color -- purple.

Also, I noticed the first cloud I made uses basic Spanish articles (such as: La, El, de, en...). So I'm ommitting them this time around.

```python
stopwords = set(STOPWORDS)
to_add = ['La','El','feat','Feat','de','en','Del','Es']
for i in to_add:
    stopwords.add(i)
```

```python
def purple_color_func(font_size, position, orientation, random_state=None,
                    **kwargs):
    hue = 280  # Purple hue value
    saturation = random.randint(70, 100)  # Random saturation between 70 and 100
    lightness = random.randint(30, 70)  # Random lightness between 30 and 70
    return "hsl({}, {}%, {}%)".format(hue, saturation, lightness)
```

```python
wordcloud = WordCloud(background_color='white', max_font_size=60, stopwords=stopwords, margin=10,
               random_state=1).generate(tracks_text)
plt.imshow(wordcloud.recolor(color_func=purple_color_func),
           interpolation="bilinear")
plt.axis("off")
plt.show()
```
